```python
import pandas as pd
from sqlalchemy import create_engine
import os
import csv
```

```python
#first file
csv_file = "Car_sales.csv"
car_sales1_df = pd.read_csv(csv_file, encoding="utf-8")
car_sales1_df.head()
```

```python
car_sales1 = car_sales1_df[["Manufacturer", "Model", "Vehicle_type", "Engine_size", "Horsepower","Wheelbase", "Width", 
                          "Length", "Curb_weight", "Fuel_capacity", "Fuel_efficiency", ]]
car_sales1
```

```python
new_car_sales1_df= car_sales1.rename(columns= 
    {"Vehicle_type": "Vehicle_Type", 
     "Engine_size": "Engine_Size", 
     "Curb_weight": "Curb_Weight", 
     "Fuel_capacity":"Fuel_Capacity", 
     "Fuel_efficiency" : "Fuel_Efficiency"
    }
                                 )
```

```python
new_car_sales1_df.insert(0,'id', range(1, 1+ len(new_car_sales1_df)), True)
```

```python
new_car_sales1_df
```

```python
new_car_sales1_df.to_csv("new_car_sales1")
```

```python
#second file
cars_df = pd.read_csv('cars.csv')
```

```python
cars_df.head()
```

```python
count = 0
wanted_columns = []
for val in cars_df.dtypes:
    if val != bool:
        wanted_columns.append(cars_df.columns[count])
    count += 1

wanted_columns
```

```python
cleaned_cars = cars_df[wanted_columns]
```

```python
cleaned_cars = cleaned_cars.drop('number_of_photos',axis = 1)
cleaned_cars = cleaned_cars.drop('engine_fuel',axis = 1)
cleaned_cars = cleaned_cars.drop('location_region',axis = 1)
```

```python
cleaned_cars = cleaned_cars.rename( columns = {
    'manufacturer_name': 'Manufacturer',
    'model_name': 'Model',
    'transmission': 'Transmission',
    'color': 'Color',
    'odometer_value': 'Odometer',
    'year_produced': 'Year',
    'engine_type': 'Engine_Type',
    'engine_capacity': 'Engine_Capacity',
    'body_type': 'Body',
    'state': 'State',
    'drivetrain': 'Drivetrain',
    'price_usd': 'Price',
    'up_counter': 'Up_Counter',
    'duration_listed': 'List_Duration'
}
                                  )
```

```python
cleaned_cars = cleaned_cars.drop('Up_Counter',axis = 1)
```

```python
cleaned_cars.insert(0,'id', range(1, 1+ len(cleaned_cars)), True)
```

```python
cleaned_cars
```

```python
cleaned_cars.to_csv('cleaned_cars.csv')
```
